## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
# TODO
import numpy as np
import pandas as pd
import random
# Suceso: la suma mayor a 7 o un número par
suceso=0
for i in range(1000):
    D1=random.randint(1,6)
    D2=random.randint(1,6)
    sum=D1+D2
    if sum>7 or sum%2==0:
        suceso +=1
        
prob=suceso/1000

print(f"La probabilidad empírica de que al tirar dos dados (1000 veces) su suma sea mayor a 7 o un número par es de {prob}.")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
ball_box = {}

# Crea la cajas con las bolas
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
experimento=[]
exp_colores=[]
#Suceso 1: que salgan 3 bolas blancas y dos rojas.
S1=0
#Suceso 2: que todas las bolas sean del mismo color.
S2=0
for i in range(1000):
    tirada=[]
    white=0
    red=0
    green=0
    colores=[]
    for j in range(5):
        S=random.randint(0,59)
        tirada.append(ball_box[S])
        if ball_box[S]=='White':
            white +=1
        elif ball_box[S]=='Red':
            red +=1
        else:
            green +=1
    if white==3 and red==2:
        S1 +=1
    if white==5 or red==5 or green==5:
        S2 +=1
    colores.append([white,red,green])
    experimento.append(tirada)
    exp_colores.append(colores)
print(experimento)
print(exp_colores)
print(f"El Suceso 1 ocurrió {S1} veces")
print(f"El Suceso 2 ocurrió {S2} veces")
probS1=S1/1000
probS2=S2/1000

print(f"La probabilidad empírica de que salgan 3 bolas blancas y dos rojas (en 1000 ejecuciones) es de {probS1}")
print(f"La probabilidad empírica de que todas las bolas sean del mismo color (en 1000 ejecuciones) es de {probS2}")
```
